### 2.2.1.対称行列、エルミート行列の対角化

- 対称行列、エルミート行列は、それぞれ直交行列、ユニタリ行列で対角化可能です。



- ここで重要なことは固有値はすべて実数であることです。対称行列の対角化は複素ベクトルを考えなくても、実数の枠内ですべて話しを行えるので、文系の方にとっては朗報で特に統計学で因子分析や主成分分析などで固有値や固有ベクトルが出てきますが扱う行列が対称行列なので複素数がでてこなくて、文系の方を悩ませないようです。　話は脱線しますが、アッティアの「数学とは何か」って本があります。M.Atiyah　って言えば、超有名な数学者、Atiyah-Singer のIndex Theorem や　K-theory またコンセビッチ
との共著のグロモフーウィッテン不変量に関する論文などなどいろいろな業績があります。この本の中で、　偉大なガウスは「$\sqrt{-1}$についての真の形而上学はかんたんなことではない」といってるといって、いかに虚数を考えるという数学者の創造性が不可能なことに挑戦したか、またそれによって、非常に多くのものを後世にもたらしかたか言っています。

- オイラーの公式とかは、当たり前のように使いますが、指数関数と三角関数がつながるというのは、また、指数関数は微分しても変わらない関数で、こんなきれいな関数を手に入れたことが、例えば、場の量子論から量子電磁力学へとゲージ対称性をふくめてゲージ場に移行する際の可換ゲージ場のゲージ群が$U(1)$ つまり $e^{i\theta}$ のなす群であることを考えると、本当になくてはならないものになってると言えるでしょう。


- しかし文系の人にとって数が複素数っていったら、???になってしまいます。まして複素ベクトルとかいったら、ベクトルで３歩さがって、複素で５歩くらい下がります。数というのは彼らにとっては大きさを表す道具ですから、順序構造が入らいない、単に代数的構造しかもたない複素数は、高校ですこーし出たけど、忘却の彼方ではないでしょうか。現在、データサイエンスや機械学習、AIと騒がれて、文系の大学でもデータサイエンスのコースを履修に置くコースが増えてきてます。文系に数学が必要になってきてますが、でも残念ながらそういう数学には複素数は出てこないのです。

- エルミート行列の対角化は、ちょうど量子力学に対応します。量子力学は、これまで数だった力学変数、運動量やエネルギー、位置や時間といったものが、作用素、平たく言えば、行列に変わります。我々が知り得るのは、観測したときの状態だけで、その量はその力学変数の固有値になります。いちおうこの固有値は実数なわけです。それはこれらの力学変数はエルミート作用素だからです。


- また重要なことは、同時対角化できるための必要十分条件は、２つの行列が可換なことです。ところが、$[x,\frac{\partial}{\partial x}]$は０でありません。また $[t,\frac{\partial}{\partial t}]$もそうです。つまり、時間とエネルギーは同時には測れない、また場所と運動量つまり速度も同時に測れません。これは従来の古典力学のような、物体の動きのイメージを全く否定してしまった。それがハイゼンベルグの不確定性関係であり、これ故に量子論はわかりにくくなります。

- もう一つ重要なこととして、正規行列があります。正規行列とは$AA^* =A^*A$ を満たす行列です。行列がユニタリ行列で対角化できるための必要十分条件は、正規行列であることです。この証明は、非常にグタグタ具体的な証明もあればスッキリと書いた証明もあって、いろいろありますので、ご自分でいろいろさがして見てください。

- もう一つの注意点は、直交行列しろユニタリ行列にしろ、固有値の異なる固有空間同士は直交し合うということです。まあ、だから固有空間の中でグラムシュミット使うことで直交行列やユニタリ行列で対角化できるわけなんですが。
